```python
import pandas as pd
```

```python
coverage = pd.read_table('coverage_filtcovgt0.tsv', header=None, names='sn rname startpos endpos numreads covbases coverage meandepth meanbaseq meanmapq'.split())
coverage
```

```python
vsinfo = pd.read_table('/jdfssz1/ST_HEALTH/P20Z10200N0206/renzirui/EA_redo/ViralSequences/EastAfrica_256viralstrains.improved.fna.lenstat.tsv', header=None, names='rname length N_count'.split())
vsinfo
```

```python
coverage = coverage.merge(vsinfo, on='rname', how='left')
coverage['non-N_length'] = coverage['length'] - coverage['N_count']
coverage
```

```python
coverage['non-N_coverage'] = coverage['covbases'] / coverage['non-N_length'] * 100  # calculate non-N coverage
coverage
```

```python
coverage_filtered = coverage[(coverage['non-N_coverage']>40)&(coverage['covbases']>1000)]
coverage_filtered
```

```python
coverage_filtered[['sn', 'rname']].to_csv('/jdfssz1/ST_HEALTH/P20Z10200N0206/renzirui/EA_redo/VirusQuantification/mismatch_calc.tsv', sep='\t', header=None, index=None)
```

```python
mismatch = pd.read_table('mismatch_calc_v2.tsv', header=None, names=['sn', 'rname', 'mismatch'])
mismatch
```

```python
merged = pd.merge(coverage_filtered, mismatch, on=['sn', 'rname'], how='left')
merged['pmismatch'] = merged['mismatch']*100
merged
```

```python
merged[merged['rname']=='MIX-ParV-19A']
```

```python
vsname2vsinfo = pd.read_excel('/jdfssz1/ST_HEALTH/P20Z10200N0206/renzirui/EA_redo/Tables/2_VSinfo.xlsx')[['ViralStrain_Name', 'Reference_Sequence','Reference_Family']]
vsname2vsinfo.columns = ['vsname', 'rname', 'vfamily']
vsname2vsinfo
```

```python
merged_addvsname = merged.merge(vsname2vsinfo)
merged_addvsname['80ani_clus'] = merged_addvsname['vsname'].apply(lambda x: x[:-1])
merged_addvsname['95ani_clus'] = merged_addvsname['vsname']
merged_addvsname
```

```python
rawdata_mismatch_gt10 = merged_addvsname[merged_addvsname['pmismatch']<10]

rmdup_result = []
for sample in rawdata_mismatch_gt10.groupby('sn'):
    [sample, dfsample] = sample
    for vfamily in dfsample.groupby('vfamily'):
    #种水平(80% ANI Cluster)
        dfvfamily = vfamily[1]
        vfamily = vfamily[0]
        print(sample, vfamily)
        dfvfamily_sorted = dfvfamily.sort_values('pmismatch')
        dfvfamily_minpervspecies = dfvfamily_sorted.drop_duplicates('80ani_clus')
        #若至少有一种最低错配率低于5%，则仅取所有错配率低于5%的株结果
        if dfvfamily_minpervspecies['pmismatch'].min() <= 6:
            spec_rmdup = dfvfamily[dfvfamily['pmismatch'] <= 6]
        #若各种最低错配率均高于5%，取覆盖度最高的种的所有株结果
        else:
            highest_coverage_spec = dfvfamily.sort_values('covbases', ascending=False).reset_index().loc[0, '80ani_clus']
            spec_rmdup = dfvfamily[dfvfamily['80ani_clus']==highest_coverage_spec]
            
        #株水平(0.95)去冗余
        #若错配率均高于2%，取覆盖度最高
        for dfspec in spec_rmdup.groupby('80ani_clus'):
            vspclus, dfspec = dfspec
            highest_coverage = dfspec.sort_values('covbases', ascending=False).reset_index().loc[0, 'covbases']
            df_hcov = dfspec[dfspec['covbases']==highest_coverage]
            if len(df_hcov) > 1:
                lowest_mismatch = df_hcov.sort_values('pmismatch', ascending=True).reset_index().loc[0, '95ani_clus']
                strain_rmdup = dfspec[dfspec['95ani_clus']==highest_coverage]
            else:
                strain_rmdup = df_hcov
            strain_rmdup = pd.concat([strain_rmdup, dfspec[dfspec['pmismatch']<1]]).drop_duplicates()
            rmdup_result.append(strain_rmdup)
            
rmdup_df = pd.concat(rmdup_result, axis=0)
```

```python
merge_cleandata = rmdup_df.merge(metadata[['sn', 'Sample_ID', 'numreads_m', 'cleandata_gb']])
merge_cleandata['RPM'] = merge_cleandata['numreads'] / merge_cleandata['numreads_m'] 
```
